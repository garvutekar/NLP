```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
b = np.array([0, 0, 0, 1])

weight = np.zeros(2)
bias = 0

for _ in range(100):
    done = True
    for i in range(len(a)):
        z = np.dot(a[i], weight) + bias
        b_pred = 1 if z >= 0 else 0
        error = b[i] - b_pred
        weight += 0.1 * error * a[i]
        bias += 0.1 * error
        if error != 0:
            done = False
    if done:
        break
```

```python
print("Weights:", weight, "Bias:", bias)

plt.scatter(a[:,0], a[:,1], c=b, s=100)
x1 = np.array([-0.5, 1.5])
x2 = -(weight[0] * x1 + bias) / weight[1]
plt.plot(x1, x2, 'k--')
plt.show()
```
