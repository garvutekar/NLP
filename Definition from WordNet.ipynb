```python
from nltk.corpus import wordnet
```

```python
def find_definitions(word):
    definitions = []
    for synset in wordnet.synsets(word):
        definitions.append(synset.definition())
    
    return definitions
```

```python
words = ["dog", "happy", "run", "beautiful", "python"]

for word in words:
    print(f"\nDefinitions for '{word}':")
    defs = find_definitions(word)
    
    if defs:
        for i, definition in enumerate(defs, 1):
            print(f"{i}. {definition}")
    else:
        print("No definitions found")
```
